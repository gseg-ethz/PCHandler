```python
import numpy as np
from scipy.spatial.transform import Rotation
```

```python
def compute_shift(arr: np.ndarray, decimal_magnitude: int = 4):
    return np.median(np.round(arr, decimals=-(decimal_magnitude - 1)), axis=0)

def is_shift_needed(arr: np.ndarray, decimal_magnitude: int = 4) -> np.bool_:
    return np.any(np.abs(arr) >= _optimal_range(decimal_magnitude))

def _optimal_range(decimal_magnitude: int = 4):
    return 10 ** decimal_magnitude
```

```python
project_origin = np.array([300_000_01.023, 1_000_010.0001, 116.00])
```

```python
socs = np.eye(4)
rot = Rotation.from_euler('zyx', [np.pi/2, 0, 0]).as_matrix()
translation = project_origin
socs[:3, :3] = rot
socs[:3, 3] = translation
```

```python
x_s = np.random.rand(10, 3)*50 + [20, -40, 6.4]
print(x_s)
```

```python
x_s_1 = np.column_stack((x_s, np.ones(x_s.shape[0])))
x_p = (socs @ x_s_1.T).T[:, :3]
shift = compute_shift(x_p)
print(shift)
```

```python
x_opt = x_p - shift
print(x_opt)
```

```python
print(f'SOCS: {is_shift_needed(x_s)}')
print(f'project_coords: {is_shift_needed(x_p)}')
print(f'optimal_coords: {is_shift_needed(x_opt)}')
```
